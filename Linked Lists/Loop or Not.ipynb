```python
class Node:
    def __init__(self, val):
        self.val = val
        self.next = None
```

```python
def insert(root, val):
    curr = root
    while(curr.next is not None):
        curr = curr.next
    curr.next = Node(val)
```

```python
def printList(root):
    curr = root
    while(curr is not None):
        print(curr.val, end="->")
        curr = curr.next
    print("None")
```

```python
root = Node(5)
```

```python
printList(root)
```

```python
def insertL(root, val):
    curr = root
    while(curr.next is not None):
        curr = curr.next
    x = Node(val)
    curr.next = x
    return x
```
