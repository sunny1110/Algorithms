```python
def fibonacci(x, map):
    print(map)
    if x in map:
        return map[x]
    if x==0:
        return 0
    if x==1:
        return 1
    else:
        result = (fibonacci((x-1), map)+fibonacci((x-2), map))
        map[x] = result
        return result
```

```python
def firstN(k):
    map = {}
    for i in range(k):
        print(fibonacci(i, map), end=" ")
```

```python
firstN(15)
```
