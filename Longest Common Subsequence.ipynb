```python
def lcsUtil(a, b, m, n, dict):
#     print(dict)
    if((m, n) in dict):
        print("I am calling the dict @ {}, {}".format(m, n))
        return dict[(m, n)]
    if m==0 or n==0:
        return ""
    
    if a[m-1]==b[n-1]:
        subLcs = lcsUtil(a, b, m-1, n-1, dict)+a[m-1]
        dict[(m, n)] = subLcs
        return subLcs
        
    else:
        bR = lcsUtil(a, b, m, n-1, dict) 
        aR = lcsUtil(a, b, m-1, n, dict)
        dict[(m, n-1)] = bR
        dict[(m-1, n)] = aR
        if(len(aR)>len(bR)):
            return aR
        else:
            return bR
```
